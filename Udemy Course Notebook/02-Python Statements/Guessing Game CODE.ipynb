```python
import random
Number = random.randint(1,100)
Guesses = [0]

# Print Staements
print("\t\t\t******WELCOME******"+'\n'+"\t\t\t  Guess The Number")
print("\nRULES:")
print("1. Your guess shouldn't be less than 1 or greater than 100")
print("2. WARM: On your first turn, if the guess is within 10 of the number")
print("3. COLD: On your first turn, if the guess is further than 10 away from the number")
print("4. WARMER: If the guess is closer to the number than the previous guess")
print("5. COLDER: If the guess is farther from the number than the previous guess")



while True:
    
    Guess = int(input('\nEnter your guess: '))
    
    if Guess <1 or Guess >100:
        print("OUT OF BOUNDS!!!\nNumber shouldn't be less then 1 or more then 100")
        continue
    
    
    # if guess is correct, tell them they've guessed correctly and how many guesses it took!   
    if Guess == Number:
        print("You Won")
        print(f"Total number of tries: {len(Guesses)}")
        break
    
    # if guess is incorrect, add guess to the list
    Guesses.append(Guess)
    
    
    # when testing the first guess, guesses[-2]==0, which evaluates to False
    # and brings us down to the second section
    
    if Guesses[-2]:
        if abs(Guesses[-1] - Number) < abs(Guesses[-2] - Number):
            print("WARMER")
        
        elif abs(Guesses[-1] - Number) > abs(Guesses[-2] - Number):
            print("COLDER")
        
    else:
        if abs(Guesses[-1] - Number) <= 10:
            print("WARM")         

        elif abs(Guesses[-1] - Number) > 10:
            print("COLD")
```
